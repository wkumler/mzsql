```python
#RETRIEVES PACKED DATA

import sqlite3

conn = sqlite3.connect("180205_Poo_TruePoo_Full1.mzDB")

# Query to select the first non-NULL entry in the `data` column from the `bounding_box` table
query = "SELECT data FROM bounding_box LIMIT 1;"

cursor = conn.cursor()
cursor.execute(query)
blob_data = cursor.fetchone()[0]  # Retrieve the first row's data

# Close the connection
conn.close()

# Print the blob data
print("First row from 'data' column in 'bounding_box':", blob_data)
```

```python
#RETRIEVES UNPACKED DATA
import sqlite3
import struct

conn = sqlite3.connect("180205_Poo_TruePoo_Full1.mzDB")

# Fetch the first BLOB entry in `data` column from `bounding_box`
cursor = conn.cursor()
cursor.execute("SELECT data FROM bounding_box LIMIT 1")
blob_data = cursor.fetchone()[0]

conn.close()

# Interpret the blob data as an array of 32-bit floats (if applicable)
float_data = struct.unpack(f'{len(blob_data) // 4}f', blob_data)

# Print the extracted float values
print(float_data)
```
