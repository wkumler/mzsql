```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pyteomics import mzml, mzmlb
import pyopenms
import pymzml
import mzsql
import timeit
```

## Spectrum access

```python
mzsql.get_spec_mzml_pyteomics("demo_data/180205_Poo_TruePoo_Full1.mzML", 1)
mzsql.get_spec_mzml_pymzml("demo_data/180205_Poo_TruePoo_Full1.mzML", 3)
mzsql.get_spec_mzml_pyopenms("demo_data/180205_Poo_TruePoo_Full1.mzML", 1)
print("empty")
```

```python
pyteomics_spec = timeit.repeat('mzsql.get_spec_mzml_pyteomics("demo_data/180205_Poo_TruePoo_Full1.mzML", 1)', globals=globals(), number=1, repeat=10)
pymzml_spec = timeit.repeat('mzsql.get_spec_mzml_pymzml("demo_data/180205_Poo_TruePoo_Full1.mzML", 3)', globals=globals(), number=1, repeat=10)
pyopenms_spec = timeit.repeat('mzsql.get_spec_mzml_pyopenms("demo_data/180205_Poo_TruePoo_Full1.mzML", 1)', globals=globals(), number=1, repeat=10)
```

```python
plt.boxplot([pyteomics_spec, pymzml_spec, pyopenms_spec], tick_labels=["pyteomics", "pymzml", "pyopenms"])
plt.axhline(y=0, color="k")
plt.show()
```
